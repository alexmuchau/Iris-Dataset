# **Adquirindo Data**

```python
data = pd.read_csv('./data/Iris.csv')
data = data.drop(columns = 'Id').drop_duplicates()

data.head()
```

```python
x = data.drop(columns = 'Species')
y = data.Species
```

# **Classificação**

### *random_forest*

```python
from sklearn.metrics import accuracy_score

def acc(y, predicts):
    acc_score = accuracy_score(y, predicts) * 100
    print("Accuracy score %.2f%%" % (acc_score))
```

```python
from sklearn.model_selection import train_test_split

SEED=123
np.random.seed(SEED)

train_x, test_x, train_y, test_y = train_test_split(x, y)
```

```python
from sklearn.ensemble import RandomForestClassifier

def random_forest_classify(train_x, test_x, train_y):
  model = RandomForestClassifier()
  model.fit(train_x, train_y)
  predicts = model.predict(test_x)
  return predicts

acc(test_y, random_forest_classify(train_x, test_x, train_y))
```

### *LinearSVC*

```python
from sklearn.svm import LinearSVC

def linear_svc_classify(train_x, test_x, train_y):
  model = LinearSVC()
  model.fit(train_x, train_y)
  predicts = model.predict(test_x)
  return predicts

acc(test_y, linear_svc_classify(train_x, test_x, train_y))
```
